#### tao_xau_nhi_phan

```python
def sinh_nhi_phan(n:int):
    if n == 1:
        return ['0', '1']
    
    c = []
    for i in sinh_nhi_phan(n-1):
        c.append('0' + i)
    for i in sinh_nhi_phan(n-1):
        c.append('1' + i)
    return c

for i in sinh_nhi_phan(3):
    print(i)
```

#### buble sort

```python
C = [5,4,3,2,1]

for i in range(len(C)-1):
    for j in range(i+1,len(C)):
        if(C[i]>C[j]):
            tmp = C[i]
            C[i] = C[j]
            C[j] = tmp

C
```

```python
def tao_hoan_vi(a):
    n=len(a)
    if(n==1):
        return [a]
    A=[]
    for p in tao_hoan_vi(a[:-1]):
        for i in range(n):
            b = p.copy()
            b.insert(i,a[-1])
            A.append(b)
    return A

tao_hoan_vi([1,2,3])
```

```python
# tinh chinh hop p(n,r)
def tinhgt(n):
    if (n==0):
        return 1
    return n*tinhgt(n-1)

def tinh_chinh_hop(n,r):
    a = tinhgt(n)/tinhgt(n-r)
    return a
tinh_chinh_hop(5,3)
```

```python
# tim chinh hop chap r phan tu cua day n phan tu

def tim_chinh_hop(a,r):
    if(r==1):
        return [[i] for i in a]
    P = []
    for i in range(len(a)):
        b = a.copy()
        x = b.pop(i)
        for p in tim_chinh_hop(b,r-1):
            p = [x] + p
            P.append(p)
    return P

print(len(tim_chinh_hop ([1,2,3,4,5] , 3)))
```

```python
# hoán vị lặp
def permutations_with_replacement(a,n):
    r = len(a)
    if sum(n) == 0:
        return [[]]
    P = []
    for i in range(r):
        if n[i] > 0:
            n_ = n.copy()
            n_[i] -= 1
            for p in permutations_with_replacement(a, n_):
                p = [a[i]] + p
                P.append(p)
    return P

permutations_with_replacement(['C','O','L'],[1,2,1])
```

```python
# có bao nhiêu tổng riêng của n
def solutions(n,r):
    if n==1:
        return [[r]]
    if r==0:
        return [[0 for _ in range(n)]]
    A=[]
    for a in solutions(n-1,r):
        A.append([0]+a)
    for a in solutions(n,r-1):
        a[0] += 1
        A.append(a)
    return A
def summands(n):
    A = []
    for k in range(1,n+1):
        for a in solutions(k,n-k):
            A.append([i+1 for i in a])
    return A
summands(10) #kết quả từ 1 đến n, phải đảo ngược nếu cần
```
